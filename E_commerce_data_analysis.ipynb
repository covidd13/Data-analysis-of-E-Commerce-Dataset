Monetary

```python
monetary = dataset.groupby('CustomerID')['Total'].sum()
monetary = pd.DataFrame(monetary).reset_index()
monetary.head()
```

Frequency

```python
freq = dataset.drop_duplicates(subset=['InvoiceNo']).groupby('CustomerID')['InvoiceNo'].count()
freq = pd.DataFrame(freq).reset_index()
freq.head()
```

Recency

```python
def recency_cerrection(recency):
    ans = str(recency).split(' ')[0]
    return(int(ans))
```

```python
dataset['Date'] = pd.to_datetime(dataset['InvoiceDate'])
```

```python
today = dataset['Date'].max()
today
```

```python
dataset['rank'] = dataset.sort_values(['CustomerID','Date']).groupby(['CustomerID'])['Date'].rank(method='min').astype(int)

recency = dataset[dataset['rank'] == 1]
recency['recency'] = pd.to_datetime('2011-12-10') - recency['Date']

recency['recency'] = recency['recency'].apply(recency_cerrection)

recency = recency[['CustomerID','recency']]
recency = recency.drop_duplicates()
recency.head()
```

```python
final_df = freq.merge(monetary, on='CustomerID').merge(recency, on='CustomerID')
final_df.columns = ['CustomerId', 'freq', 'monetary', 'recency']
final_df.head()
```

```python
final_df.shape
```

```python
final_df.sort_values(['recency'])
```

```python
def get_group(Q1, Q2, Q3, value):
    if value <= Q1:
        return '1'
    elif Q1 < value <= Q2:
        return '2'
    elif Q2 < value <= Q3:
        return '3'
    elif Q3 < value:
        return '4'
```

Compute Q1,Q2 and Q3 for frequency

```python
freq_Q1, freq_Q2, freq_Q3 = final_df['freq'].quantile([0.25,0.5,0.75])
```

Getting group for each frequency

```python
freq_group = []

for i in range(len(final_df)):
    freq = final_df.iloc[i]['freq']
    freq_group.append(get_group(freq_Q1, freq_Q2, freq_Q3, freq)) 

final_df['freq_group'] = freq_group
```

Compute Q1,Q2 and Q3 for recency

```python
rece_Q1, rece_Q2, rece_Q3 = final_df['recency'].quantile([0.25,0.5,0.75])
```

Getting group for each recency

```python
recency_group = []

for i in range(len(final_df)):
    recency = final_df.iloc[i]['recency']
    recency_group.append(get_group(rece_Q1, rece_Q2, rece_Q3, recency)) 

final_df['recency_group'] = recency_group
```

Compute Q1,Q2 and Q3 for monetary

```python
mone_Q1, mone_Q2, mone_Q3 = final_df['monetary'].quantile([0.25,0.5,0.75])
```

Getting group for each monetary

```python
monetary_group = []

for i in range(len(final_df)):
    monetary = final_df.iloc[i]['monetary']
    monetary_group.append(get_group(mone_Q1, mone_Q2, mone_Q3, monetary)) 

final_df['monetary_group'] = monetary_group
```

Comuputing RFM score of each customer

```python
final_df['rfm'] = final_df['recency_group'] + final_df['freq_group'] + final_df['monetary_group']
```

```python
final_df.head()
```

```python
def get_group(rfm):
    if rfm == '144':
        return 'Best'
    elif rfm == '344':
        return 'Almost Lost'
    elif rfm == '444':
        return 'Lost Big Spenders'
    elif rfm == '441':
        return 'Lost Cheap'
    elif rfm[1] == '4':
        return 'Loyal'
    elif rfm[2] == '4':
        return 'Big Spenders'
    else:
        return 'Normal'
```

```python
final_df['Segment'] = final_df['rfm'].apply(get_group)
```
